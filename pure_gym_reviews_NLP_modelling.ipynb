```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re


# Define preprocessing function (clean text)
def preprocess_text(text):
    if not isinstance(text, str):  # Ensure text is a string
        return ""

    # Convert text to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    clean_words = [word for word in text.split() if word not in stop_words]

    # Return the cleaned text as a string
    return " ".join(clean_words)
```

```python
# Define tokenization function
def tokenize_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    return tokens
```

```python
# Load the pipeline and tokenizer
emotion_pipeline = pipeline(
    "text-classification",
    model="bhadresh-savani/bert-base-uncased-emotion",
    tokenizer="bhadresh-savani/bert-base-uncased-emotion"
)

# Load the tokenizer for manual truncation
tokenizer = AutoTokenizer.from_pretrained("bhadresh-savani/bert-base-uncased-emotion")
```

```python
def extract_emotion(text, pipeline):
    # Ensure the input is valid
    if not isinstance(text, str) or not text.strip():
        return {"label": "neutral", "score": 0.0}  # Default for invalid inputs

    try:
        # Run the emotion pipeline on the raw text
        result = pipeline(text)
        # Extract the highest scoring emotion
        top_emotion = max(result, key=lambda x: x['score'])
        return top_emotion
    except Exception as e:
        # Log errors if any
        print(f"Error processing text: {repr(text)} - Exception: {str(e)}")
        return {"label": "error", "score": 0.0}  # Default for failed inputs
```

```python
# Apply to Google dataset
google_data['Emotion'] = google_data['Comment'].apply(lambda x: extract_emotion(x, emotion_pipeline))

# Apply to Trustpilot dataset
trustpilot_data['Emotion'] = trustpilot_data['Comment'].apply(lambda x: extract_emotion(x, emotion_pipeline))
```

```python
# Load the Falcon-7B-Instruct model
text_generation_pipeline = pipeline(
    "text-generation",  # Task type
    model="microsoft/Phi-3.5-mini-instruct",  # Model name
    device=0  # Use GPU if available; set to -1 for CPU
)

# Configure max length for each review
max_length = 1000  # Set the max token length
```

```python
# Define the prompt format
def format_prompt(review):
    return f"In the following customer review, pick out the main 3 topics. Return them in a numbered list format, with each one on a new line.\n\nReview: {review}"

# Subset reviews (e.g., 5 for testing)
subset_reviews = negative_google_data['Comment'].head(5).tolist()  # Adjust the number as needed
prompts = [format_prompt(review) for review in subset_reviews]

# Process prompts in smaller batches
batch_size = 2  # Adjust batch size to fit GPU memory
results = []

for i in range(0, len(prompts), batch_size):
    batch_prompts = prompts[i:i + batch_size]
    try:
        batch_responses = text_generation_pipeline(
            batch_prompts,
            max_length=256,  # Reduce max length for memory optimization
            num_return_sequences=1,
            truncation=True  # Explicit truncation
        )
        # Append results
        results.extend([response[0]['generated_text'] for response in batch_responses])
    except RuntimeError as e:
        print(f"Error processing batch {i // batch_size + 1}: {e}")
        break

# Print results
for i, (review, result) in enumerate(zip(subset_reviews, results)):
    print(f"Review {i + 1}:\n{review}\n")
    print(f"Generated Topics:\n{result}\n")
    print("-" * 50)
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Define preprocessing function
def preprocess_text(text):
    if not isinstance(text, str):  # Ensure the input is a string
        return ""
    # Convert to lowercase
    text = text.lower()
    # Remove numbers and special characters
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize and remove stopwords
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)  # Rejoin tokens into a cleaned string

# Apply preprocessing to the combined reviews
preprocessed_reviews_top_30 = combined_reviews_top_30.apply(preprocess_text)
```
